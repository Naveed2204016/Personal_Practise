```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('concrete_data.csv')
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X=df.drop('Strength',axis=1)
Y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
np.mean(cross_val_score(lr,X,Y,scoring='r2'))
```

```python
for cols in X.columns:
    plt.figure(figsize=(14,4))

    plt.subplot(121)
    sns.distplot(X[cols])
    plt.title(cols)

    plt.subplot(122)
    stats.probplot(X[cols],dist="norm",plot=plt)
    plt.show()
```

```python

pt=PowerTransformer(method='box-cox')

x_train_transformed=pt.fit_transform(x_train+0.0001)
x_test_transformed=pt.transform(x_test+0.0001)

pd.DataFrame({'cols':x_train.columns,'box cox lambdas':pt.lambdas_})
```

```python
pt=PowerTransformer(method='box-cox')
x_trans=pt.fit_transform(X+0.0000001)
lr=LinearRegression()
np.mean(cross_val_score(lr,x_trans,Y,cv=10,scoring='r2'))
```

```python
pt=PowerTransformer()
x_transformed2=pt.fit_transform(X)
lr=LinearRegression()
np.mean(cross_val_score(lr,x_transformed2,Y,cv=10,scoring='r2')) 
```
