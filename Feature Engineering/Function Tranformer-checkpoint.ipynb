```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Titanic-Dataset.csv',usecols=['Age','Fare','Survived'])
```

```python
df.sample(5)
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
X=df.iloc[:,1:]
Y=df.iloc[:,0:1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=LogisticRegression()
clf2=DecisionTreeClassifier()
clf.fit(x_train,y_train)
clf2.fit(x_train,y_train)
y_pred=clf.predict(x_test)
y_pred2=clf2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("LT",accuracy_score(y_test,y_pred))
print("DT",accuracy_score(y_test,y_pred2))
```

```python
import scipy.stats as stats
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
```

```python
plt.figure(figsize=(14,4))

plt.subplot(121)
sns.distplot(x_train['Age'])
plt.title('Age')

plt.subplot(122)
stats.probplot(x_train['Age'],dist="norm",plot=plt)
plt.title('Age')
plt.show()
```

```python
plt.figure(figsize=(14,4))

plt.subplot(121)
sns.distplot(x_train['Fare'])
plt.title('Fare')

plt.subplot(122)
stats.probplot(x_train['Fare'],dist="norm",plot=plt)
plt.title('Fare')
plt.show()
```

```python
pt=FunctionTransformer(func=np.log1p)
```

```python
x_train_transformed=pt.fit_transform(x_train)
```

```python
x_test_transformed=pt.transform(x_test)
```

```python
clf.fit(x_train_transformed,y_train)
clf2.fit(x_train_transformed,y_train)
y_pred_trans=clf.predict(x_test_transformed)
y_pred2_trans=clf2.predict(x_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
print("LT",accuracy_score(y_test,y_pred_trans)*100)
print("DT",accuracy_score(y_test,y_pred2_trans)*100)
```

```python
cross_val_score(clf,x_train_transformed,y_train,cv=10,scoring='accuracy').mean()
```

```python
plt.figure(figsize=(14,4))

plt.subplot(121)
stats.probplot(x_train['Fare'],dist="norm",plot=plt)
plt.title('Fare')

plt.subplot(122)
stats.probplot(x_train_transformed['Fare'],dist="norm",plot=plt)
plt.title('Fare')
plt.show()
```

```python
tf=ColumnTransformer(transformers=[('log',FunctionTransformer(func=np.log1p),['Fare'])],remainder='passthrough')
```

```python
x_train_transformed_final=tf.fit_transform(x_train)
```

```python
cross_val_score(clf,x_train_transformed_final,y_train,cv=10,scoring='accuracy').mean()
```

```python
def play(transform):
    X=df.iloc[:,1:]
    Y=df.iloc[:,0:1]
    tf=ColumnTransformer(transformers=[('func',FunctionTransformer(transform),['Fare'])],remainder='passthrough')
    x_trans=tf.fit_transform(X)
    clf=LogisticRegression()
    print("accuracy",np.mean(cross_val_score(clf,x_trans,Y,cv=10,scoring='accuracy')))
    plt.figure(figsize=(14,4))

    plt.subplot(121)
    stats.probplot(X['Fare'],dist="norm",plot=plt)
    plt.title('Fare')

    plt.subplot(122)
    stats.probplot(x_trans[:,0],dist="norm",plot=plt)
    plt.title('Fare')
    plt.show()
```
