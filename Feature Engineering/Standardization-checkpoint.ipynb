```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
X=df.drop('Purchased',axis=1)
Y=df.drop(['Age','EstimatedSalary'],axis=1)
```

<h1>Train test split</h1>

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

<h1>StandardScaler</h1>

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train_s=scaler.transform(x_train)
x_test_s=scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train_s=pd.DataFrame(x_train_s,columns=x_train.columns)
x_test_s=pd.DataFrame(x_test_s,columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_s.describe(),1)
```

<h1>Effect of Scaling</h1>

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(x_train_s['Age'],x_train_s['EstimatedSalary'],color='red')
ax2.set_title('After Scaling')
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)
ax2.set_title('After Scaling')
sns.kdeplot(x_train_s['Age'],ax=ax2)
sns.kdeplot(x_train_s['EstimatedSalary'],ax=ax2)
plt.show()
```

<h1>Comparisons of Distributions</h1>

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'],ax=ax1)
ax2.set_title('After Scaling')
sns.kdeplot(x_train_s['Age'],ax=ax2)
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)
ax2.set_title('After Scaling')
sns.kdeplot(x_train_s['EstimatedSalary'],ax=ax2)
plt.show()
```

<h1>Why scaling is important?</h1>

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
model2=LogisticRegression()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train_s,y_train)
```

```python
y_pred=model1.predict(x_test)
y_pred_s=model2.predict(x_test_s)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_s))
```

```python
from sklearn.tree import DecisionTreeClassifier
f1=DecisionTreeClassifier()
f2=DecisionTreeClassifier()
```

```python
f1.fit(x_train,y_train)
f2.fit(x_train_s,y_train)
```

```python
y_pred1=f1.predict(x_test)
y_pred_s1=f2.predict(x_test_s)
```

```python
print("Actual",accuracy_score(y_test,y_pred1))
print("Scaled",accuracy_score(y_test,y_pred_s1))
```
