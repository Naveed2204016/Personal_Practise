```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class label', 'Alcohol', 'Malic acid']
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict={1:'red',3:'green',2:'blue'}
sns.scatterplot(data=df,x='Alcohol',y='Malic acid',hue=df['Class label'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
X=df.drop('Class label',axis=1)
Y=df.drop(['Alcohol', 'Malic acid'],axis=1)
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
ob1=MinMaxScaler()
ob1.fit(x_train)
x_train_n=ob1.transform(x_train)
x_test_n=ob1.transform(x_test)
```

```python
x_train_n=pd.DataFrame(x_train_n,columns=x_train.columns)
x_test_n=pd.DataFrame(x_test_n,columns=x_train.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_n.describe(),1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Alcohol'], x_train['Malic acid'])
ax1.set_title("Before Scaling")
ax2.scatter(x_train_n['Alcohol'], x_train_n['Malic acid'])
ax2.set_title("After Scaling")
plt.show()
```

```python
ig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'], ax=ax1)
sns.kdeplot(x_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_n['Alcohol'], ax=ax2)
sns.kdeplot(x_train_n['Malic acid'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Alcohol Distribution Before Scaling')
sns.kdeplot(x_train['Alcohol'], ax=ax1)

# after scaling
ax2.set_title('Alcohol Distribution After Standard Scaling')
sns.kdeplot(x_train_n['Alcohol'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Malic Acid Distribution Before Scaling')
sns.kdeplot(x_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('Malic Acid Distribution After Standard Scaling')
sns.kdeplot(x_train_n['Malic acid'], ax=ax2)
plt.show()
```
